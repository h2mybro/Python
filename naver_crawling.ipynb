```python
#크롤링을 통한 각각의 기능모음 만들기
#1.오늘의 날씨
import requests
from bs4 import BeautifulSoup

def make_soup(url):
    headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"}
    res=requests.get(url,headers=headers)
    res.raise_for_status() #정상연결유무체크함수
    soup=BeautifulSoup(res.text,"lxml")
    return soup

# 오늘의 날씨!!!
def crawl_weather():
    print("[오늘의 날씨!!!]")
    print()
    url="https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%84%9C%EC%9A%B8%EB%82%A0%EC%94%A8"
#     res=requests.get(url)
#     res.raise_for_status()#연결유무에 대한 정상체크
#     soup=BeautifulSoup(res.text,"lxml")
    soup=make_soup(url)

    cast=soup.find("p",{"class":"summary"}).get_text()
    curr_temp=soup.find("div",{"class":"temperature_text"}).get_text()
    min_temp=soup.find("span",{"class":"lowest"}).get_text()
    max_temp=soup.find("span",{"class":"highest"}).get_text()
    rain_fall=soup.find("div",{"class":"cell_weather"}).get_text()
    print(cast)
    print(curr_temp)
    print(min_temp)
    print(max_temp)
    print(rain_fall)

#오늘의 주요뉴스
def crawl_headline():
    print()
    print("[오늘의 주요뉴스!!!]")
    url="https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=100"
#     res=requests.get(url)
#     res.raise_for_status()#연결유무에 대한 정상체크
#     soup=BeautifulSoup(res.text,"lxml")
    soup=make_soup(url)
    ul=soup.find("ul",{"class":"section_list_ranking_press _rankingList"})
    lis=ul.find_all("li")
   
    for li in lis:
        div=li.find("div",{"class":"list_text_inner"})
        title=div.get_text()
        a_tag=div.find("a").get('href')
        
        print(title, end="")
        print(a_tag)

# 오늘의 주식!!!
def crawl_stock():
    print()
    print("[오늘의 주식!!!]")
    print()
    url = "https://finance.naver.com/"
    soup = make_soup(url)
    ul = soup.find("div",{"class":"section_strategy"})
    lis = ul.find_all("li")
    
    for li in lis:
        span = li.find("span").get_text()
        a_tag = li.find("a").get('href')
        
        print(span)
        print("https://finance.naver.com{}" .format(a_tag))
        print()
        
# 오늘의 스포츠!!!
def crawl_sport():
    print()
    print("[오늘의 스포츠!!!]")
    print()
    url = "https://sports.news.naver.com/index"
    soup = make_soup(url)
    ul = soup.find("ul",{"class":"today_list"})
    lis = ul.find_all("li")
    
    for li in lis:
        div = li.find("div", {"class":"text_area"}).find("strong", {"class":"title"})
        title = div.get_text()
        a_tag = li.find("a").get('href')
        
        print(title, end = "")
        print()
        print("https://sports.news.naver.com{}" .format(a_tag))
        print()
        
# 오늘의 부동산!!!
def crawl_realty():
    print()
    print("[오늘의 부동산!!!]")
    print()
    url = "https://land.naver.com/"
    soup = make_soup(url)
    ul = soup.find("div",{"class":"news NE=a:tdn"})
    lis = ul.find_all("li")
    
    for li in lis:
        title = li.find("a").get('title')
        a_tag = li.find("a").get('href')
        
        print(title)
        print("https://land.naver.com{}" .format(a_tag))
        print()
        
if __name__=='__main__':#현재 프로그램내용이 진행되는 프로그램이니?
    crawl_weather()#오늘의 날씨정보
    crawl_headline() #오늘의 주요뉴스
    crawl_stock() # 오늘의 주식
    crawl_sport() # 오늘의 스포츠
    crawl_realty() # 오늘의 부동산

```
